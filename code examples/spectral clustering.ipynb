```python
from sklearn.datasets import make_moons, make_circles
from sklearn.cluster import KMeans


datasets = [(2,make_moons(1000, noise=.1, random_state=4)[0]), 
            (3, np.r_[make_circles(500, factor=.6, noise=.04, random_state=4)[0],
                      make_circles(500, factor=.05, noise=.02, random_state=4)[0]]),
            (3,z)]
predictions = [KMeans(k).fit_predict(X) for k, X in datasets]


fig = make_subplots(rows=1, cols=3, row_heights=[200])
for i,((k,X),pred) in enumerate(zip(*[datasets, predictions])):
    fig.add_trace(go.Scatter(x=X[:,0], y=X[:,1], mode="markers", marker=dict(color=pred, colorscale=class_colors(k), line=dict(color="black", width=1)), showlegend=False), row=1, col=i+1)

fig = fig.update_layout(height=200, width=600).update_xaxes(showticklabels=False).update_yaxes(showticklabels=False)

fig
```
